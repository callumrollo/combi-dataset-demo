```python
import xarray as xr
import matplotlib.pyplot as plt
```

Three datasets are all from same origin data. **legato** uses CTD as base timebase. **oxygen** uses oxygen sensor as timebase. **non_blank** retains data from both sensors & the flourometer

```python
legato = xr.open_dataset('data/legato.nc')
oxygen = xr.open_dataset('data/oxygen.nc')
non_blank = xr.open_dataset('data/non_blank.nc')
```

```python
!du -h data/*.nc
```

```python
offset = 10
var = 'temperature'
fig, ax = plt.subplots(figsize=(12,6))
scat_kwarg = {'s': 5}
ax.scatter(legato.time, legato[var], label='legato', **scat_kwarg)
ax.scatter(oxygen.time, oxygen[var] - 1 * offset, label='oxygen', **scat_kwarg)
ax.scatter(non_blank.time, non_blank[var] - 2 * offset, label='non-blank', **scat_kwarg)
ax.set(ylabel = var)
ax.legend(loc=3)
```

Note how the oxygen ds misses most of the legato data

```python
offset = 40
var = 'oxygen_concentration'
fig, ax = plt.subplots(figsize=(12,6))
scat_kwarg = {'s': 5}
ax.scatter(legato.time, legato[var], label='legato', **scat_kwarg)
ax.scatter(oxygen.time, oxygen[var] -1 * offset, label='oxygen', **scat_kwarg)
ax.scatter(non_blank.time, non_blank[var] - 2 * offset, label='non-blank', **scat_kwarg)
ax.set(ylabel = var)
ax.legend(loc=3)
```

The legato ds misses much of the oxygen data. non-blank has data from both datasets
